```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# Selecting the numeric columns for clustering
X = crop_production_df[['ph_level', 'organic_matter (%)', 'nitrogen_content (kg/ha)', 
          'phosphorus_content (kg/ha)', 'potassium_content (kg/ha)']]

# Standardizing the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Perform K-Means clustering with 3 clusters
kmeans = KMeans(n_clusters=3, random_state=42)
crop_production_df['Cluster'] = kmeans.fit_predict(X_scaled)

# Plot clusters
sns.scatterplot(x='ph_level', y='nitrogen_content (kg/ha)', hue='Cluster', data=crop_production_df, palette='deep')
plt.title('Clusters of Soil Samples based on pH and Nitrogen Content')
plt.show()
```

```python
#predictive modelling (linear regression)
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# Feature variables (using all the numeric features except the one to predict)
x = crop_production_df[['ph_level', 'organic_matter (%)', 'phosphorus_content (kg/ha)', 
          'potassium_content (kg/ha)']]

# Target variable: Nitrogen Content
y = crop_production_df['nitrogen_content (kg/ha)']

# Split the data into train and test sets
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

# Fit the model
model = LinearRegression()
model.fit(x_train, y_train)

# Predict
y_pred = model.predict(x_test)

# Evaluate the model
mse = mean_squared_error(y_test, y_pred)
print("Mean Squared Error:", mse)

# Compare predicted vs actual
plt.stem(y_test, y_pred)
plt.xlabel("Actual Nitrogen Content")
plt.ylabel("Predicted Nitrogen Content")
plt.title("Actual vs Predicted Nitrogen Content")
plt.show()
```

```python
#Determining the Fertility index of the project
crop_production_df['fertility_index'] = (crop_production_df['organic_matter (%)']*0.4
                  + crop_production_df['nitrogen_content (kg/ha)']*0.3
                 + crop_production_df['phosphorus_content (kg/ha)']*0.2
                  + crop_production_df['potassium_content (kg/ha)']*0.1)
fertility_ranking = crop_production_df[['District','soil_type','fertility_index']].sort_values(by='fertility_index',ascending=False)
fertility_ranking
```
